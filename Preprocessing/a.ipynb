```python
columns_to_split = ['SIG.STR.', 'TOTAL STR.', 'HEAD', 'BODY', 'LEG']

for col in columns_to_split:
    col_index = df.columns.get_loc(col)
    
    new_cols = df[col].str.split(' of ', expand=True)
    new_cols.columns = [f'{col}_Success', f'{col}_Attempt']

    df.drop(columns=[col], inplace=True)

    for i, new_col in enumerate(new_cols.columns):
        df.insert(col_index + i, new_col, new_cols[new_col])

print(df.head())
```
